```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-np.pi,np.pi,200)
y=np.sin(x)
plt.plot(x,y)
plt.title('Graph of sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,200)
y=np.cos(x)
plt.plot(x,y)
plt.title('Graph of cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,200)#200 is no.of parts
y=(np.sin(x))/(np.cos(x))
plt.plot(x,y)
plt.title('Graph of tan(x)')
plt.xlabel('x')
plt.ylabel('tan(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,200)
y=1/np.sin(x)
plt.plot(x,y)
plt.title('Graph of cosec(x)')
plt.xlabel('x')
plt.ylabel('cosec(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,200)
y=1/np.cos(x)
plt.plot(x,y)
plt.title('Graph of sec(x)')
plt.xlabel('x')
plt.ylabel('sec(x)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,200)
y=1/np.tan(x)
plt.plot(x,y)
plt.title('Graph of cot(x)')
plt.xlabel('x')
plt.ylabel('cot(x)')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Amazon Sales Data.csv",encoding='unicode_escape')
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
df.info()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df
```

```python
import seaborn as sns
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  sns.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum()
```

```python
print(sales_gen)
```

```python
#total amount for all states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum()
```

```python
print(sales_state)
```

```python
#sorting
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(ascending=False,by='Amount').head()
sales_state.head()
sns.barplot(x="State",y="Amount",data=sales_state)
```
